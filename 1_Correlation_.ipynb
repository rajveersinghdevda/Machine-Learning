```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
file_path = '/content/drive/MyDrive/Data set2/processed.cleveland.data'  # Replace with your file path
df = pd.read_csv(file_path, header=None)
df.columns = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'num']
```

```python

# Replace '?' with NaN
df = df.replace('?', np.nan)

```

```python
# Handle NaN values (choose one of the methods above)
# Example: Impute with mean for 'ca' and 'thal' columns
df['ca'] = pd.to_numeric(df['ca'], errors='coerce')
df['thal'] = pd.to_numeric(df['thal'], errors='coerce')
df['ca'].fillna(df['ca'].mean(), inplace=True)
df['thal'].fillna(df['thal'].mean(), inplace=True)
```

```python

# Calculate the correlation matrix
corr_matrix = df.corr()
```

```python

# Create a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Heart Disease Dataset')
plt.show()
```
